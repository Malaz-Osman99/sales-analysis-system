```python
import pandas as pd

df = pd.read_excel("C:/Users/thela/OneDrive/Desktop/project/sales_analytics_system/data/raw/sales_dataset_raw.xlsx")
df["order_date"] = pd.to_numeric(df["order_date"], errors="coerce")
df["order_date"] = pd.to_datetime(df["order_date"], origin="1899-12-30", unit="D")
df["order_date"] = df["order_date"].dt.date
df["order_date"].head(10)
```

```python
df = df.dropna(subset=["order_date"])
df["order_date"].isnull().sum()
```

```python
df["total_price"] = df["total_price"].astype(str)

# حذف أي فواصل أو مسافات
df["total_price"] = df["total_price"].str.replace(",", "", regex=False)
df["total_price"] = df["total_price"].str.replace(" ", "", regex=False)

# تحويله لرقم
df["total_price"] = pd.to_numeric(df["total_price"], errors="coerce")
df["total_price"].describe()
```

```python
# نخلي أي قيمة سالبة = NaN
df.loc[df["total_price"] < 0, "total_price"] = None
df.loc[df["total_price"] > 10000, "total_price"] = df["total_price"] / 1000
df["total_price"].describe()
```

```python
#تأكد من نوع البيانات + تنظيف لو فيها فواصل/رموز
df["quantity"] = df["quantity"].astype(str)
df["quantity"] = df["quantity"].str.replace(",", "", regex=False)
df["quantity"] = df["quantity"].str.replace(" ", "", regex=False)
df["quantity"] = pd.to_numeric(df["quantity"], errors="coerce")
df["quantity"].describe()
```

```python
df[df["quantity"] <= 0]
df.loc[df["quantity"] <= 0, "quantity"] = None
df["quantity"].isnull().sum()
df["quantity"] = df["quantity"].round().astype("Int64")
df["quantity"].head(10)
```

```python
summary = pd.DataFrame({
    "Column": df.columns,
    "DataType": df.dtypes.values,
    "Missing": df.isnull().sum().values,
    "Unique": df.nunique().values
})
summary
```

```python
df["order_date"] = pd.to_datetime(df["order_date"], errors="coerce")

df["order_date"].dtype
text_cols = ["Product Name", "Category", "customer_type", "City"]

df[text_cols] = df[text_cols].fillna("Unknown")
df["order_id"] = pd.to_numeric(df["order_id"], errors="coerce").astype("Int64")
df = df.dropna(subset=["order_id"])
df = df.dropna(subset=["quantity", "total_price"])
```

```python
df["City"] = df["City"].str.strip()
df["Category"] = df["Category"].str.strip()
df["customer_type"] = df["customer_type"].str.strip()
df["payment_method"] = df["payment_method"].str.strip()
df["City"] = df["City"].str.title()
df["Category"] = df["Category"].str.title()
df["customer_type"] = df["customer_type"].str.title()
df["payment_method"] = df["payment_method"].str.title()
```

```python
# ================================
# Standardization (Text Cleaning)
# ================================

# 1) إزالة المسافات الزائدة من بداية/نهاية النصوص
text_cols = ["City", "Category", "customer_type", "payment_method", "Product Name"]
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()

# 2) توحيد حالة الأحرف (اختياري)
# هنا نخلي أول حرف كبير لكل كلمة (مناسب للمدن والأنواع)
df["City"] = df["City"].str.title()
df["Category"] = df["Category"].str.title()
df["customer_type"] = df["customer_type"].str.title()
df["payment_method"] = df["payment_method"].str.title()

# 3) تصحيح أخطاء التهجئة وتوحيد القيم (Mapping)
# City fixes
city_mapping = {
    "Los Angles": "Los Angeles",
    "San Franciso": "San Francisco"
}
df["City"] = df["City"].replace(city_mapping)

# Payment method fixes
payment_mapping = {
    "Ewallet": "E-Wallet",
    "E Wallet": "E-Wallet",
    "Cc": "Card",
    "Bt": "Bank Transfer"
}
df["payment_method"] = df["payment_method"].replace(payment_mapping)

# Customer type fixes
customer_mapping = {
    "Company": "Business",
    "B2B": "Business"
}
df["customer_type"] = df["customer_type"].replace(customer_mapping)

# 4) تحويل القيم "Nan" النصية إلى NaN الحقيقي (احتياط مهم)
df.replace("Nan", pd.NA, inplace=True)
df.replace("None", pd.NA, inplace=True)

# ================================
# Check results after standardization
# ================================
print("Payment Methods:\n", df["payment_method"].value_counts(), "\n")
print("Customer Types:\n", df["customer_type"].value_counts(), "\n")
print("Cities (Top 20):\n", df["City"].value_counts().head(20), "\n")
```
